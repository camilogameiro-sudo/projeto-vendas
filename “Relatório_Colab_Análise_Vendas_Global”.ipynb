```python
import pandas as pd
import os

BASE_PATH = "/content/drive/MyDrive/13.Final project"

print("Arquivos encontrados na pasta base:")
print(os.listdir(BASE_PATH))

products_path = os.path.join(BASE_PATH, "products.csv")
events_path = os.path.join(BASE_PATH, "events.csv")
countries_path = os.path.join(BASE_PATH, "countries.csv")

products = pd.read_csv(products_path, low_memory=False)
events = pd.read_csv(events_path, low_memory=False)
countries = pd.read_csv(countries_path, low_memory=False)

print("Products:\n", products.head(), "\n")
print("Events:\n", events.head(), "\n")
print("Countries:\n", countries.head(), "\n")


```

Aqui carregamos os dados de produtos, vendas e países. Vamos analisar e limpar essas tabelas antes de qualquer conclusão.

```python
def resumo_df(df, nome):
    print(f"\n--- {nome} ---")
    print("Tamanho:", df.shape)
    print("\nColunas:", list(df.columns))
    print("\nTipos de dados:")
    print(df.dtypes)
    print("\nEstatísticas gerais (numéricas):")
    display(df.describe())
    print("\nValores ausentes por coluna:")
    display(df.isna().sum())

resumo_df(products, "Products")
resumo_df(events, "Events")
resumo_df(countries, "Countries")
```

```python
import pandas as pd
import numpy as np

def clean_columns(df):
    df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
    return df

products = clean_columns(products)
events = clean_columns(events)
countries = clean_columns(countries)

events['order_date'] = pd.to_datetime(events['order_date'], errors='coerce')
events['ship_date'] = pd.to_datetime(events['ship_date'], errors='coerce')


if 'country_code' in events.columns:
    events['country_code'] = events['country_code'].fillna('Unknown')

events = events[events['units_sold'].notna()]

for col in ['alpha-2', 'region', 'sub-region']:
    if col in countries.columns:
        countries[col] = countries[col].fillna('Unknown')

def normalize_string(s):
    if pd.isna(s):
        return s
    s = str(s).strip().upper()
    return s

if 'country_code' in events.columns:
    events['country_code'] = events['country_code'].apply(normalize_string)

if 'alpha-2' in countries.columns:
    countries['alpha-2'] = countries['alpha-2'].apply(normalize_string)

products.drop_duplicates(inplace=True)
events.drop_duplicates(inplace=True)
countries.drop_duplicates(inplace=True)

events['revenue'] = events['units_sold'] * events['unit_price']
events['cost_total'] = events['units_sold'] * events['unit_cost']
events['profit'] = events['revenue'] - events['cost_total']

print("Products:")
display(products.head())

print("Events:")
display(events.head())

print("Countries:")
display(countries.head())

print("Shapes após limpeza:")
print("Products:", products.shape)
print("Events:", events.shape)
print("Countries:", countries.shape)
```

Limpamos as tabelas, padronizamos nomes, preenchemos dados ausentes e calculamos receita e lucro de cada pedido.

```python
df = events.merge(products, left_on='product_id', right_on='id', how='left')

df = df.merge(countries, left_on='country_code', right_on='alpha-2', how='left')

df.drop(columns=['id', 'alpha-2'], inplace=True)

print("Shape do DataFrame final:", df.shape)
display(df.head())
```

Agora temos um único dataset, contendo todas as informações necessárias para análise.

```python
total_orders = df['order_id'].nunique()
total_revenue = df['revenue'].sum()
total_profit = df['profit'].sum()
total_countries = df['country_code'].nunique()
total_products = df['product_id'].nunique()

print(f"Total de pedidos: {total_orders}")
print(f"Receita total: ${total_revenue:,.2f}")
print(f"Lucro total: ${total_profit:,.2f}")
print(f"Número de países: {total_countries}")
print(f"Número de produtos: {total_products}")
```

Esses números nos dão uma visão geral do tamanho e desempenho da empresa.

```python
import matplotlib.pyplot as plt
import seaborn as sns

revenue_by_category = df.groupby('item_type')['revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=revenue_by_category.index, y=revenue_by_category.values)
plt.title("Receita por Categoria de Produto")
plt.ylabel("Receita")
plt.xticks(rotation=45)
plt.show()

revenue_by_channel = df.groupby('sales_channel')['revenue'].sum()
plt.figure(figsize=(6,4))
sns.barplot(x=revenue_by_channel.index, y=revenue_by_channel.values)
plt.title("Receita por Canal de Vendas")
plt.ylabel("Receita")
plt.show()
```

Conclusão: Podemos ver quais categorias geram mais receita e se o canal online ou offline é mais lucrativo.

```python
revenue_by_country = df.groupby('country_code')['revenue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=revenue_by_country.index, y=revenue_by_country.values)
plt.title("Top 10 Países por Receita")
plt.ylabel("Receita")
plt.show()

revenue_by_region = df.groupby('region')['revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=revenue_by_region.index, y=revenue_by_region.values)
plt.title("Receita por Região")
plt.ylabel("Receita")
plt.xticks(rotation=45)
plt.show()
```

Conclusão: Identificamos os países e regiões mais lucrativos, ajudando a priorizar mercados estratégicos.

```python
df['lead_time'] = (df['ship_date'] - df['order_date']).dt.days

lead_by_category = df.groupby('item_type')['lead_time'].mean().sort_values()
plt.figure(figsize=(10,6))
sns.barplot(x=lead_by_category.index, y=lead_by_category.values)
plt.title("Lead Time Médio por Categoria")
plt.ylabel("Dias")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='lead_time', y='profit')
plt.title("Lucro vs Lead Time")
plt.xlabel("Lead Time (dias)")
plt.ylabel("Lucro")
plt.show()
```

Podemos ver se produtos com envio mais rápido têm impacto no lucro.

```python
df['order_month'] = df['order_date'].dt.to_period('M')
monthly_revenue = df.groupby('order_month')['revenue'].sum()

plt.figure(figsize=(12,6))
monthly_revenue.plot()
plt.title("Receita Mensal ao Longo do Tempo")
plt.ylabel("Receita")
plt.xlabel("Mês")
plt.show()
```

dentificamos tendências de crescimento ou queda nas vendas ao longo do tempo.

```python
df['order_day'] = df['order_date'].dt.day_name()
day_revenue = df.groupby('order_day')['revenue'].sum().reindex([
    'Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])

plt.figure(figsize=(10,5))
sns.barplot(x=day_revenue.index, y=day_revenue.values)
plt.title("Receita por Dia da Semana")
plt.ylabel("Receita")
plt.show()
```

odemos perceber quais dias da semana geram mais vendas e se há produtos sazonais.
